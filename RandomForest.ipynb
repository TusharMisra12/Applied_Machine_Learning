```python
#20BHI10028 - TUSHAR MISRA
import pandas as pd
from pandas import read_csv
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
```

```python
df = pd.read_csv("data3.csv")
df.head()
```

```python
X = df.drop(['city'],axis='columns')
y = df['city']

from sklearn.model_selection import train_test_split
```

```python
X
```

```python
y
```

```python
test_size_arr=[0.1,0.2,0.5,0.9]
result=[]
for i in test_size_arr:
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=i)
    from sklearn.ensemble import RandomForestClassifier
    model = RandomForestClassifier(n_estimators=20)
    model.fit(X_train, y_train)
    y_predicted = model.predict(X_test)
    accuracy=model.score(X_test, y_test)
    result+=accuracy,
print(result)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.ylabel("Accuracy Score")
plt.xlabel("Test Size")
plt.title("20BHI10028 - Accuracy Score vs Test Size")
plt.bar(test_size_arr,result,width=0.02,color="green")
plt.show
```
